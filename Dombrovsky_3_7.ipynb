```python
import pandas as pd
import chardet
import numpy as np
import matplotlib.pyplot as plt

file_path = 'C:\\Users\\wsluvv\\Downloads\\pract3\\Global YouTube Statistics.csv'
with open(file_path, 'rb') as file:
    result = chardet.detect(file.read())

print(result['encoding'])

data = pd.read_csv(file_path, encoding='latin-1')
print(data.head())
data.info()
```

```python
data = data.rename(columns={
    "Номер": "rank",
    "Ютубер": "Youtuber",
    "Пiдписники": "subscribers",
    "Кiлькiсть переглядiв": "video views",
    "Категорiя": "category",
    "Назва": "Title",
    "Завантаження": "uploads",
    "Країна": "Country",
    "Абревіатура": "Abbreviation",
    "Тип каналу": "channel_type",
    "Підписалися_за_останні_30_днів": "subscribers_for_last_30_days",
    "Рiк створення": "created_year",
    "Мiсяць створення": "created_month",
    "День створення": "created_date",
    "Валовий охоплення вищою освітою (%)": "Gross tertiary education enrollment (%)",
    "Населення": "Population",
    "Рівень безробіття": "Unemployment rate",
    "Міське_населення": "Urban_population",
    "Широта": "Latitude",
    "Довгота": "Longitude"
    })
data = data.replace("—", np.nan)
data
```

```python
unique_countries_count = data['Country'].nunique()
print("Кiлькiсть унiкальних країн:", unique_countries_count)
```

```python
data['video views'].plot(kind='hist', bins=20, edgecolor='k', color='r')

plt.title('Розподiл преглядiв на Youtube')
plt.xlabel('Кiлькiсть переглядiв')
plt.ylabel('Частота')


plt.show()
```

```python
#Максимальна кількість переглядів
min_number_of_views = data['video views'].min()
year_min_number_of_views = data.loc[data['video views'] == min_number_of_views, 'Youtuber'].values[0]
print(f"Мінімальна кількість переглядів - {min_number_of_views} була на каналі {year_min_number_of_views}.")
#Максима кількість переглядів
max_number_of_views = data['video views'].max()
year_max_number_of_views = data.loc[data['video views'] == max_number_of_views, 'Youtuber'].values[0]
print(f"Максимальна кількість переглядів - {max_number_of_views} була на каналі {year_max_number_of_views}.")
```

```python
country_video_counts = data.groupby('Country')['uploads'].count()
country_with_most_videos = country_video_counts.idxmax()
print("Країна з найбільшою кількістю відео на YouTube:", country_with_most_videos)
```

```python
max_uploads_index = data['uploads'].idxmax()
video_with_max_uploads = data.loc[max_uploads_index, 'Title']
min_uploads_index = data['uploads'].idxmin()
video_with_min_uploads = data.loc[min_uploads_index, 'Title']
print("Відео з найменшою категорією 'uploads':", video_with_min_uploads)
print("Відео з найбільшою категорією 'uploads':", video_with_max_uploads)
```
